```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_a_integrar(x):
    return x**2
a=2
b=5
N=1000
x=np.random.uniform(a,b,N)
y=np.random.uniform(0,funcion_a_integrar(b),N)
```

```python
n_bajo_curva=np.sum(y < funcion_a_integrar(x))
n_bajo_curva
```

```python
respuesta=(b-a)* funcion_a_integrar(b)*n_bajo_curva/N
respuesta
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,100),funcion_a_integrar(np.linspace(a,b,100)),'r')
plt.show()
```

```python
# sacar el pi del montecarlo
# si tienes un cuadrado de 2x2 y un radio de un circulo de 1 
def funcion_a_integrar(x):
    return np.sqrt(1-x**2)
a=0
b=1
N=1000000
x=np.random.uniform(a,b,N)
y=np.random.uniform(0,1,N)

n_bajo_curva=np.sum(y < funcion_a_integrar(x))
respuesta= 4*n_bajo_curva/N
respuesta
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,100),funcion_a_integrar(np.linspace(a,b,100)),'r')
plt.show()
```
